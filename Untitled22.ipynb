```python
!pip install textstat spacy matplotlib seaborn scikit-learn
!python -m spacy download en_core_web_sm

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
import re
from collections import Counter
import textstat
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import string

nlp = spacy.load('en_core_web_sm')

print("=== ЛАБОРАТОРНАЯ РАБОТА ПО АНАЛИЗУ ТЕКСТОВ ===")
print("Студенческий билет: 1032249820")
print("Выбранные тексты: Story 9, Story 8, Story 2")
print()

texts = {}

texts[9] = """All children, except one, grow up. They soon know that they will grow up, and the way Wendy knew was this. One day when she was two years old she was playing in a garden, and she plucked another flower and ran with it to her mother. I suppose she must have looked rather delightful, for Mrs. Darling put her hand to her heart and cried, "Oh, why can't you remain like this for ever!" This was all that passed between them on the subject, but henceforth Wendy knew that she must grow up. You always know after you are two. Two is the beginning of the end.

Of course they lived at 14, and until Wendy came her mother was the chief one. She was a lovely lady, with a romantic mind and such a sweet mocking mouth. Her romantic mind was like the tiny boxes, one within the other, that come from the puzzling East, however many you discover there is always one more; and her sweet mocking mouth had one kiss on it that Wendy could never get, though there it was, perfectly conspicuous in the right-hand corner."""

texts[8] = """To Sherlock Holmes she is always the woman. I have seldom heard him mention her under any other name. In his eyes she eclipses and predominates the whole of her sex. It was not that he felt any emotion akin to love for Irene Adler. All emotions, and that one particularly, were abhorrent to his cold, precise but admirably balanced mind. He was, I take it, the most perfect reasoning and observing machine that the world has seen, but as a lover he would have placed himself in a false position.

He never spoke of the softer passions, save with a gibe and a sneer. They were admirable things for the observer—excellent for drawing the veil from men's motives and actions. But for the trained reasoner to admit such intrusions into his own delicate and finely adjusted temperament was to introduce a distracting factor which might throw a doubt upon all his mental results. Grit in a sensitive instrument, or a crack in one of his own high-power lenses, would not be more disturbing than a strong emotion in a nature such as his."""

texts[2] = """It was a queer, sultry summer, the summer they electrocuted the Rosenbergs, and I didn't know what I was doing in New York. I'm stupid about executions. The idea of being electrocuted makes me sick, and that's all there was to read about in the papers—goggle-eyed headlines staring up at me on every street corner and at the fusty, peanut-smelling mouth of every subway. It had nothing to do with me, but I couldn't help wondering what it would be like, being burned alive all along your nerves.

I thought it must be the worst thing in the world. New York was bad enough. By nine in the morning the fake, country-wet freshness that somehow seeped in overnight evaporated like the tail end of a sweet dream. Mirage-gray at the bottom of their granite canyons, the hot streets wavered in the sun, the car tops sizzled and glittered, and the dry, cindery dust blew into my eyes and down my throat."""

print(f"Загружено текстов: {len(texts)}")
print()

print("=" * 80)
print("1. БАЗОВЫЙ АНАЛИЗ ТЕКСТОВ")
print("=" * 80)

results = {}
for text_id, text in texts.items():
    print(f"\n--- Текст {text_id} ---")

    char_count = len(text)
    print(f"Количество символов: {char_count}")

    words = text.split()
    word_count = len(words)
    print(f"Количество слов: {word_count}")

    sentences = re.split(r'[.!?]+', text)
    sentences = [s.strip() for s in sentences if s.strip()]
    sentence_count = len(sentences)
    print(f"Количество предложений: {sentence_count}")

    words_clean = [re.sub(r'[^\w\s]', '', word) for word in words]
    longest_words = sorted(set(words_clean), key=len, reverse=True)[:3]
    print(f"Три самых длинных слова: {longest_words}")

    words_per_sentence = [len(re.findall(r'\b\w+\b', s)) for s in sentences]
    min_words = min(words_per_sentence)
    max_words = max(words_per_sentence)
    avg_words = sum(words_per_sentence) / len(words_per_sentence)

    print(f"Минимальное количество слов в предложении: {min_words}")
    print(f"Максимальное количество слов в предложении: {max_words}")
    print(f"Среднее количество слов в предложении: {avg_words:.2f}")

    results[text_id] = {
        'text': text,
        'words': words,
        'sentences': sentences,
        'words_per_sentence': words_per_sentence
    }

print("\n" + "=" * 80)
print("2. ОЦЕНКА ЧИТАБЕЛЬНОСТИ")
print("=" * 80)

readability_scores = {}
for text_id, text in texts.items():
    print(f"\n--- Текст {text_id} ---")

    flesch_reading = textstat.flesch_reading_ease(text)
    print(f"Индекс легкости чтения Флеша: {flesch_reading:.2f}")

    if flesch_reading >= 90:
        interpretation = "Очень легко (5 класс)"
    elif flesch_reading >= 80:
        interpretation = "Легко (6 класс)"
    elif flesch_reading >= 70:
        interpretation = "Довольно легко (7 класс)"
    elif flesch_reading >= 60:
        interpretation = "Стандартный (8-9 класс)"
    elif flesch_reading >= 50:
        interpretation = "Довольно сложный (10-12 класс)"
    elif flesch_reading >= 30:
        interpretation = "Сложный (студенты)"
    else:
        interpretation = "Очень сложный (выпускники)"
    print(f"Интерпретация: {interpretation}")

    gunning_fog = textstat.gunning_fog(text)
    print(f"Индекс туманности Ганнинга: {gunning_fog:.2f}")

    readability_scores[text_id] = {
        'flesch': flesch_reading,
        'gunning_fog': gunning_fog
    }

best_flesch = max(readability_scores.items(), key=lambda x: x[1]['flesch'])
best_gunning = min(readability_scores.items(), key=lambda x: x[1]['gunning_fog'])

print(f"\nРЕЗУЛЬТАТЫ ОЦЕНКИ ЧИТАБЕЛЬНОСТИ:")
print(f"Самый читабельный текст по Флешу: Текст {best_flesch[0]} ({best_flesch[1]['flesch']:.2f})")
print(f"Самый читабельный текст по Ганнингу: Текст {best_gunning[0]} ({best_gunning[1]['gunning_fog']:.2f})")

print("\n" + "=" * 80)
print("3. ЧАСТОТНЫЙ АНАЛИЗ ТОКЕНОВ")
print("=" * 80)

for text_id, data in results.items():
    print(f"\n--- Текст {text_id} ---")

    word_freq = Counter(data['words'])
    top_5_tokens = word_freq.most_common(5)

    print("Пять наиболее часто встречающихся токенов:")
    for i, (token, freq) in enumerate(top_5_tokens, 1):
        print(f"{i}. '{token}': {freq} раз")

print("\n" + "=" * 80)
print("4. ЛЕММАТИЗАЦИЯ И ВИЗУАЛИЗАЦИЯ")
print("=" * 80)

plt.figure(figsize=(15, 10))

for i, (text_id, text) in enumerate(texts.items(), 1):
    doc = nlp(text)
    lemmas = [token.lemma_.lower() for token in doc if token.is_alpha and not token.is_stop]
    lemma_lengths = [len(lemma) for lemma in lemmas]

    plt.subplot(2, 3, i)
    plt.hist(lemma_lengths, bins=15, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Текст {text_id}\nРаспределение длин лемм', fontsize=12)
    plt.xlabel('Длина леммы')
    plt.ylabel('Частота')
    plt.grid(True, alpha=0.3)

    plt.subplot(2, 3, i+3)
    length_counts = Counter(lemma_lengths)
    lengths = list(length_counts.keys())
    frequencies = list(length_counts.values())

    plt.bar(lengths, frequencies, color='lightcoral', alpha=0.7)
    plt.title(f'Текст {text_id}\nЧастоты длин лемм', fontsize=12)
    plt.xlabel('Длина леммы')
    plt.ylabel('Частота')
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "=" * 80)
print("5. ОЧИСТКА ОТ СТОП-СЛОВ И НЕАЛФАВИТНЫХ СИМВОЛОВ")
print("=" * 80)

clean_results = {}
for text_id, text in texts.items():
    print(f"\n--- Текст {text_id} ---")

    doc = nlp(text)

    clean_tokens = [
        token.text.lower() for token in doc
        if not token.is_stop and token.is_alpha and not token.is_punct
    ]

    original_count = len([token for token in doc if token.is_alpha])
    clean_count = len(clean_tokens)

    print(f"Исходное количество токенов: {original_count}")
    print(f"Количество токенов после очистки: {clean_count}")
    print(f"Удалено токенов: {original_count - clean_count}")

    clean_results[text_id] = {
        'original_tokens': original_count,
        'clean_tokens': clean_count,
        'tokens_removed': original_count - clean_count,
        'clean_token_list': clean_tokens
    }

print("\n" + "=" * 80)
print("6. РАЗМЕТКА ЧАСТЕЙ РЕЧИ")
print("=" * 80)

pos_results = {}
plt.figure(figsize=(15, 10))

for i, (text_id, text) in enumerate(texts.items(), 1):
    print(f"\n--- Текст {text_id} ---")

    doc = nlp(text)
    pos_tags = [token.pos_ for token in doc if token.is_alpha]
    pos_counts = Counter(pos_tags)

    print("Количество токенов по частям речи:")
    for pos, count in pos_counts.most_common():
        pos_name = {
            'NOUN': 'Существительные',
            'VERB': 'Глаголы',
            'ADJ': 'Прилагательные',
            'ADV': 'Наречия',
            'PRON': 'Местоимения',
            'DET': 'Детерминаторы',
            'ADP': 'Предлоги',
            'CONJ': 'Союзы',
            'NUM': 'Числительные',
            'PART': 'Частицы'
        }.get(pos, pos)
        print(f"  {pos_name}: {count}")

    plt.subplot(2, 3, i)
    labels = list(pos_counts.keys())
    values = list(pos_counts.values())

    plt.bar(labels, values, color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc', '#c2c2f0', '#ffb3e6', '#c4e17f'])
    plt.title(f'Текст {text_id}\nРаспределение частей речи', fontsize=12)
    plt.xlabel('Части речи')
    plt.ylabel('Количество')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)

    plt.subplot(2, 3, i+3)
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title(f'Текст {text_id}\nДоли частей речи', fontsize=12)

    pos_results[text_id] = pos_counts

plt.tight_layout()
plt.show()

print("\n" + "=" * 80)
print("7. ИМЕНОВАННЫЕ СУЩНОСТИ")
print("=" * 80)

ner_results = {}
for text_id, text in texts.items():
    print(f"\n--- Текст {text_id} ---")

    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    entity_counts = Counter([ent.label_ for ent in doc.ents])

    print(f"Общее количество именованных сущностей: {len(entities)}")
    print("Распределение по типам:")

    entity_type_names = {
        'PERSON': 'Персоны',
        'ORG': 'Организации',
        'GPE': 'Геополитические объекты',
        'LOC': 'Локации',
        'DATE': 'Даты',
        'TIME': 'Время',
        'MONEY': 'Деньги',
        'PERCENT': 'Проценты',
        'CARDINAL': 'Числа',
        'NORP': 'Национальности/религии/политические группы'
    }

    for entity_type, count in entity_counts.most_common():
        type_name = entity_type_names.get(entity_type, entity_type)
        print(f"  {type_name}: {count}")

    print("Примеры сущностей:")
    for entity, etype in entities[:5]:
        type_name = entity_type_names.get(etype, etype)
        print(f"  '{entity}' - {type_name}")

    ner_results[text_id] = {
        'total_entities': len(entities),
        'entity_counts': entity_counts,
        'entities_list': entities
    }

print("\n" + "=" * 80)
print("8. TF-IDF ВЕКТОРИЗАЦИЯ")
print("=" * 80)

text_list = [texts[9], texts[8], texts[2]]
text_names = ['Текст 9', 'Текст 8', 'Текст 2']

vectorizer = TfidfVectorizer(
    max_features=100,
    stop_words='english',
    lowercase=True,
    ngram_range=(1, 2)
)

tfidf_matrix = vectorizer.fit_transform(text_list)
feature_names = vectorizer.get_feature_names_out()

print(f"Форма матрицы TF-IDF: {tfidf_matrix.shape}")
print(f"Количество текстов: {tfidf_matrix.shape[0]}")
print(f"Количество признаков: {tfidf_matrix.shape[1]}")
print("\nПримеры признаков (первые 20):")
print(feature_names[:20])

print("\n" + "=" * 80)
print("9. КОСИНУСНОЕ СХОДСТВО")
print("=" * 80)

cosine_sim = cosine_similarity(tfidf_matrix)

print("Матрица косинусного сходства:")
print("       Текст9  Текст8  Текст2")
for i, row in enumerate(cosine_sim):
    print(f"Текст{[9,8,2][i]}  {row[0]:.3f}   {row[1]:.3f}   {row[2]:.3f}")

max_similarity = -1
most_similar_pair = (0, 0)

for i in range(len(cosine_sim)):
    for j in range(i + 1, len(cosine_sim)):
        if cosine_sim[i][j] > max_similarity:
            max_similarity = cosine_sim[i][j]
            most_similar_pair = ([9, 8, 2][i], [9, 8, 2][j])

print(f"\nРЕЗУЛЬТАТЫ АНАЛИЗА СХОДСТВА:")
print(f"Наиболее похожие тексты: Текст {most_similar_pair[0]} и Текст {most_similar_pair[1]}")
print(f"Степень сходства: {max_similarity:.4f}")

plt.figure(figsize=(8, 6))
sns.heatmap(cosine_sim,
            xticklabels=['Текст 9', 'Текст 8', 'Текст 2'],
            yticklabels=['Текст 9', 'Текст 8', 'Текст 2'],
            annot=True, cmap='YlOrRd', fmt='.3f')
plt.title('Матрица косинусного сходства текстов')
plt.tight_layout()
plt.show()

print("\n" + "=" * 80)
print("ИТОГОВАЯ СВОДКА")
print("=" * 80)

print(f"Проанализировано текстов: 3")
print(f"Выбранные тексты: 9, 8, 2 (по номеру студенческого билета 1032249820)")
print(f"Самый читабельный текст: Текст {best_flesch[0]}")
print(f"Наиболее похожие тексты: Текст {most_similar_pair[0]} и Текст {most_similar_pair[1]}")
print(f"Общее количество слов во всех текстах: {sum(len(results[tid]['words']) for tid in [9,8,2])}")
print(f"Общее количество предложений: {sum(len(results[tid]['sentences']) for tid in [9,8,2])}")

print("\n" + "=" * 80)
print("ЛАБОРАТОРНАЯ РАБОТА ЗАВЕРШЕНА")
print("=" * 80)
```
